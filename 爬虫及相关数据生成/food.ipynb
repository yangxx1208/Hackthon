```python
query_name = "美食"
tag_name_all = ["中餐厅","外国餐厅","小吃快餐店","蛋糕甜品店","咖啡厅","茶座","酒吧"]


def dic_dic(dic):
    dict_ex = {}
    for item in dic.keys():
        if not isinstance(dic[item],(dict)):
            dict_ex[item] = dic[item]
        else: 
            dict_ex =  dict(dict_ex, **dic_dic(dic[item]))
    return dict_ex

def df_jiudian(url):
    response = requests.get(url)
    content  = response.json()
    dic_ser = dic_dic(content["results"][len(content["results"])-1])
    df = pd.DataFrame.from_dict(dic_ser,orient='index').T
    for x in range(len(content["results"])-1):
        dic_ser = dic_dic(content["results"][x])
        series = pd.DataFrame.from_dict(dic_ser,orient='index').T
        df = df.append(series)
    return df

    
df_all = df

print(df_all)


for tag_name in tag_name_all:
    print(tag_name)
    url = "http://api.map.baidu.com/place/v2/search?query={query}&tag={tag}&region=%E4%B8%8A%E6%B5%B7&scope=2&output=json&ak=qFpM7c6t6lmwp5h97uYBpcjiPA8P0fLe&page_size=20&page_num={page_num}".format(query = query_name,tag = tag_name,page_num = 0)
    response = requests.get(url)
    content = response.json()
    total = content["total"]
    pages = int(total/20)
    page_num_name_all = list(range(pages))  
    print(page_num_name_all)
    for page_num_name in page_num_name_all:
        url = "http://api.map.baidu.com/place/v2/search?query={query}&tag={tag}&region=%E4%B8%8A%E6%B5%B7&scope=2&output=json&ak=qFpM7c6t6lmwp5h97uYBpcjiPA8P0fLe&page_size=20&page_num={page_num}".format(query = query_name,tag = tag_name,page_num = page_num_name)
        print(url)
        df_df1 = df_jiudian(url)
        print(df_df1)
        df_all = pd.concat([df_all , df_df1])
        print(df_all.shape)
```
